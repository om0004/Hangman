```python
import random
file=open("wordlist.TXT")
f=[]
f1=[]
for word in file:
    f.append(word[:-1])
attempts=int(input("How many incorrect attempts do u want[1-25]:"))
word_length=int(input("What minimum word length do u want?[2-13]:"))
for word in f:
    if len(word)==word_length:
        f1.append(word)
word=random.choice(f1)       
print("Selecting a word....")
vowel="aeiou"
incorrect=0
count=0
true_length=0
hidden=[]
correct=[x for x in word]
for x in word:
    if x in vowel:
        hidden.append(x) 
    else:
        hidden.append("*")
        true_length=true_length+1
#hidden=[str("*") for x in word if x not in['a','e','i','o','u']]
won=False
previous=''
while attempts!=0:
    if count==true_length:
        print("Congratulations you guessed the word!")
        won=True
        break
    print("word:"+''.join(hidden))
    print("Attempts remaining:"+str(attempts))
    print("Previous guesses:"+previous)
    character=input("Choose the next letter:")
    if character in correct:
        for i in range(len(correct)):
            if correct[i]==character:
                hidden[i]=character
                count=count+1
        print(character+" is in the word!")            
    else:
        print(character+" is not in word")
        attempts=attempts-1
        previous=character
hidden.clear()
correct.clear()        
if won==False:
    print("sorry u lost the word was "+word)
    
    


```
